This problem is a classfication problem. So after we got the price, we will made it into 4 labels:
1. small rise (price increase within 1 standard deviation)
2. small drop (prcie drop within 1 std)
3. big rise (price rise out of 1 std)
4. big drop (price drop out of 1 std)
<br>And one problem here is what std to use. Should we use std for the whole period? I would suggest we use the std for each month or use a Moving average std to measure volatility. For simplicity, we can just use std for that month first to have some results. We can improve it later.

```python
import pandas as pd
```

```python
price = pd.read_csv("btc_price.csv")
```

## Label the price with regard to the standard deviation on a rolling base(window size is arbitrarily defined):


The labels can be named as string or number

```python
def label_price(price, windowsize):
    size = []
    for i in range(windowsize):
        size += ['nolabel']
    for i in range(windowsize, price.shape[0]):
        std = price['change'][i-windowsize:i].std()
        if price['change'][i] > 0:
            if price['change'][i] >= std:
                size += ['big rise'] #[1]
            else:
                size += ['small rise'] #[2]
        else:
            if abs(price['change'][i]) >= std:
                size += ['big drop'] #[3]
            else:
                size += ['small drop'] #[4]
    price['label'] = pd.DataFrame({'label':size})

label_price(price, 30)
```
