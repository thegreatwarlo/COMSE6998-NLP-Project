```python
Feature = pd.DataFrame()
for i in range(1,8):
    var = "features{}".format(i)
    Feature = pd.concat([Feature,eval(var)])
```

```python
Feature.index = np.arange(Feature.shape[0])
```

```python
def filter1(time):
    a, b = time.split(' ')
    year, month, day = a.split('-')
    hour,minute,second = b.split(':')
    return datetime.datetime(int(year),int(month),int(day),int(hour),int(minute),int(second))

def filter2(time):
    a, b = time.split(' ')
    year, month, day = a.split('-')
    hour,minute = b.split(':')
    return datetime.datetime(int(year),int(month),int(day),int(hour),int(minute))
```

```python
Feature.daily_discussion_date = Feature.daily_discussion_date.apply(lambda x: filter1(x))
price.time = price.time.apply(lambda x: filter2(x))
```
